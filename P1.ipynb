Below are some helper functions to help get you started. They should look familiar from the lesson!

```python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)

    line_img = np.zeros([img.shape[0],img.shape[1],3], dtype=np.uint8)
    draw_lines(line_img, lines)
    return (line_img, lines)

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
```

```python
def annotate(image):
    
    gau = gaussian_blur(image, 5)
    cannydata = canny(gau[:,:,0], 120, 150) + canny(gau[:,:,1], 120, 150) + canny(gau[:,:,2], 120, 150)
    cap = region_of_interest(cannydata, np.array([[[0,image.shape[0]],
                                                    [image.shape[1]*0.45,image.shape[0]*0.6], 
                                                    [image.shape[1]*0.55,image.shape[0]*0.6],
                                                    [image.shape[1], image.shape[0]] ] ], dtype=np.int32))
    l = cv2.HoughLinesP(cap, 2, np.pi/180, 30,np.array([]), minLineLength=20 , maxLineGap=20)
    
    line_img = np.zeros(image.shape, dtype=np.uint8)

    try:
        x1 = (l[:,0,0])
        y1 = (l[:,0,1])
        x2 = (l[:,0,2])
        y2 = (l[:,0,3])
    except:
        print("Did not find any line returning CAP")
        return cap

    sep = ((y1-y2)/(x1-x2)) 

    lineR = l[ sep > 0.4 ].flatten()
    lineL = l[ sep < -0.4 ].flatten()

    try:
        lineR = lineR.reshape(int(len(lineR)/2),2)
        [vx,vy,x,y] = cv2.fitLine(lineR,cv2.DIST_L2,0,0.01,0.01)
        l1 = (image.shape[0] - y)/vy
        l2 = (min(lineR[:,0]) - x)/vx
        cv2.line(line_img, (x+l2*vx,y+l2*vy), (x+l1*vx,y+l1*vy), [255, 0, 0], 10)    
    except:
        print("Cannot find right lane")


    try:
        lineL = lineL.reshape(int(len(lineL)/2),2)
        [vx,vy,x,y] = cv2.fitLine(lineL,cv2.DIST_L2,0,0.01,0.01)
        l1 = (image.shape[0] - y)/vy
        l2 = (max(lineL[:,0]) - x)/vx 

        cv2.line(line_img, (x+l2*vx,y+l2*vy), (x+l1*vx,y+l1*vy), [255, 0, 0], 10)
    except:
        print("Cannot find left lane")
    
    return weighted_img(line_img, image)
import os
for filename in os.listdir("test_images/"):
    try:
        image = mpimg.imread("test_images/"+filename)


        image = annotate(image)
        plt.imshow(image)
        plt.show()
    except OSError:
        print("Ignored %s" % filename)
```

# My Pipeline

### Filter Pipeline
My pipeline starts with a blur to reduce the sensitivity to small minor details, I found that this is not strictly necessary and good acceptable results can also be obtained without if the parameters of the rest of the pipeline are adjusted accordingly.

Next step is the Canny filter. Originally I made the image black and white first, but that decreases the performance of the challenge, since the grey value of the yellow line and the concrete is the same. Therefore, I now use 3 Canny filters, one for each color channel and add the result together.

The next step is using only the part of the image where the lanes are. Here there are some quite strong assumptions as to where the camera is pointing and where the horizon will be. It will likely lead to bad results for different camera mounting.

Finally I get the lines by means of Hough transform. 

### Line Averaging
To turn the lines returned by Hough into 2 two lane markers I take the following steps:
1. Separate left and right lane by calculating the slope of the lines and also filter horizontal lines which are likely not what I am interested in. This approach is a little bit crude and some sort of clustering method (similar to k-means?) according to angles might be more robust. The my current way is likely to fail for steep turns for example, but in that case there are probably other problems as well (fitting lines to steep turns does not seem to be a great Idea to beginn with).
2. After separation I do a line fit using all the points. This could probably improved by weighting longer lines higher (instead of just using endpoints once regardless of length) but it works well even without that, so I did not change it.
3. Finally I draw the line from the bottom to the point which is furthest away on the x axis.

### Potential failures:
- Steep turns will not work (but those are not lines)
- The algorithm can be confused by other lines (can be caused by shadows see extra where that happens in very few frames) in the field of the image that I am using if they have a plausible slope (not horizontal).
- If there is a large distance between short markers no line can be found (happens in extra for few frames)

Along with the improvement possibilities already outlined a major improvement could probably be achieved by using a few frames (which might be the only way of dealing with frames where almost no marker is visible). However, that reduces the flexibility of the algorithm so I did not implement it.
